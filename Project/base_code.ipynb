```python
import networkx as nx
import matplotlib.pyplot as plt
import csv
import pandas as pd
import os

G = nx.DiGraph()

with open("E:/Confirmation Bias/csv file/edges.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        G.add_edge(int(row["u"]), int(row["v"]))


with open("E:/Confirmation Bias/csv file/nodes1.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        node = int(row["node"])
        threshold = float(row["threshold"])
        opinion = row["opinion"].strip() if row["opinion"] else None

        G.add_node(node)
        G.nodes[node]["threshold"] = threshold
        G.nodes[node]["opinion"] = opinion


opinion_colors = {
    "A": "red",
    "B": "blue",
    "C": "green",
    None: "lightgray"
}

opinion_weight = {
    "A": 0.9,
    "B": 0.6,
    "C": 0.4
}

ALPHA = 0.6
BETA = 0.3
GAMMA = 0.4


def resistance(node, new_op, prev_op):
    if prev_op is None or prev_op == new_op:
        return 0.0
    return 1.0


def directed_multi_opinion_step(G):
    prev_opinion = {n: G.nodes[n]["opinion"] for n in G.nodes()}
    updates = {}

    for v in G.nodes():
        incoming = list(G.predecessors(v))
        if not incoming:
            continue

        influence = {}
        for u in incoming:
            op = prev_opinion[u]
            if op is not None:
                influence[op] = influence.get(op, 0) + 1

        for op in influence:
            influence[op] /= len(incoming)

        scores = {}
        for op, infl in influence.items():
            scores[op] = (
                ALPHA * infl +
                BETA * opinion_weight.get(op, 0) -
                GAMMA * resistance(v, op, prev_opinion[v])
            )

        valid_ops = [
            op for op, s in scores.items()
            if s >= G.nodes[v]["threshold"]
        ]

        if valid_ops:
            updates[v] = max(valid_ops, key=lambda o: scores[o])

    return updates, prev_opinion


def export_to_excel(G, iteration, prev_opinion,
                    out_dir="E:/Confirmation Bias/updateexcel"):
    os.makedirs(out_dir, exist_ok=True)

    node_rows = []
    for n in G.nodes():
        node_rows.append({
            "node": n,
            #"opinion": G.nodes[n]["opinion"],
            #"previous_opinion": prev_opinion.get(n),
            "opinion_changed": prev_opinion.get(n) != G.nodes[n]["opinion"],
            "threshold": G.nodes[n]["threshold"],
            "iteration": iteration
        })

    pd.DataFrame(node_rows).to_excel(
        f"{out_dir}/nodes_iter_{iteration}.xlsx",
        index=False
    )

    edge_rows = [{"source": u, "target": v} for u, v in G.edges()]
    pd.DataFrame(edge_rows).to_excel(
        f"{out_dir}/edges_iter_{iteration}.xlsx",
        index=False
    )

pos = nx.spring_layout(G, seed=42)

def plot_iteration(G, iteration):
    colors = [opinion_colors[G.nodes[n]["opinion"]] for n in G.nodes()]

    labels = {
        n: f"{n}\n{G.nodes[n]['opinion']}"
        for n in G.nodes()
    }

    plt.figure(figsize=(8, 8))
    nx.draw(
        G, pos,
        node_color=colors,
        node_size=800,
        edge_color="gray",
        arrows=True
    )
    nx.draw_networkx_labels(G, pos, labels, font_size=8)
    plt.title(f"Iteration {iteration}")
    plt.show()
MAX_ITER=10
for i in range(1, MAX_ITER + 1):
    updates, prev_opinion = directed_multi_opinion_step(G)

    real_changes = {
        node: op for node, op in updates.items()
        if prev_opinion[node] != op
    }

    print(f"Iteration {i}: {len(real_changes)} real opinion changes")

    if not real_changes:
        print("No new opinion changes. Converged.")
        export_to_excel(G, i, prev_opinion)
        plot_iteration(G, i)
        break

    for node, op in real_changes.items():
        G.nodes[node]["opinion"] = op

    export_to_excel(G, i, prev_opinion)
    plot_iteration(G, i)
```
